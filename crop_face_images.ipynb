```python
# Define the size of the images after resizing
size = (160, 160)
```

```python
def get_input(prompt):
    input_widget = widgets.Text(placeholder=prompt)
    display(input_widget)
    return input_widget
```

```python
# Define a function to label images with a GUI
def label_face(image):
    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    print("Entering here")
    # Wait for a key press
    input_widget = get_input("Input the name of the person in the image")
    key = input_widget.value

    # Create a disctionary for the person's images if it doesn't exist
    person_dir = os.path.join('test_images_formatted', key)
    if not os.path.exists(person_dir):
        os.makedirs(person_dir)

    # Resize the image to a fixed size and save it to the person's directory
    resized_image = cv2.resize(image, size)
    filename = os.path.join(person_dir, f'{len(os.listdir(person_dir))}.jpg')
    cv2.imwrite(filename, resized_image)
    cv2.waitKey(0)
```

```python
## 1. Collect preprocess the dataset## 1. Collect preprocess the dataset# Loop over all images files in a directory
for filename in os.listdir():
    print(f'image name: {filename}')
    # Open the image file
    image = cv2.imread(os.path.join('', filename))

    if image is None:
      print('Image file is not in a format that cv2.imread() can read.')
      continue

    else:
      print(image)
      # Detect faces in the image using a face detection model
      face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
      gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
      faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

      # Label each detected face
      for (x, y, w, h) in faces:
          face_image = image[y:y+h, x:x+w]
          label_face(face_image)
```
